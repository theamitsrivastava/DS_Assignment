```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv("C:\\ExcelR\\Data Science Assignments\\10-Recommendation system\\book.csv", encoding = "ISO-8859-1")
```

```python
book
```

```python
book = book.rename({'User.ID': 'User'}, axis=1)
```

```python
book = book.rename({'Book.Title': 'Title'}, axis=1)
```

```python
book = book.rename({'Book.Rating': 'Rating'}, axis=1)
```

```python
book
```

```python
#number of unique users in the dataset
len(book.User.unique())
```

```python
len(book.Title.unique())
```

```python
piv_book = book.pivot_table(index='User',
                                 columns='Title',
                                 values='Rating').reset_index(drop=True)
```

```python
piv_book
```

```python
#Impute those NaNs with 0 values
piv_book.fillna(0, inplace=True)
```

```python
piv_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
sim_book = 1 - pairwise_distances( piv_book.values,metric='cosine')
```

```python
sim_book
```

```python
#Store the results in a dataframe
sim_book_df = pd.DataFrame(sim_book)
```

```python
sim_book_df.index = book.User.unique()
sim_book_df.columns = book.User.unique()
```

```python
sim_book_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(sim_book, 0)
sim_book_df.iloc[0:50, 0:50]
```

```python
#Most Similar Users
sim_book_df.idxmax(axis=1)[0:50]
```
