# DBSCAN Clustering

```python
#Crime Data DBSCAN Clustering
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
crime_data = pd.read_csv("C:\\ExcelR\Data Science Assignments\\7-Clustering\\crime_data.csv")
crime_data.head()
```

```python
df = crime_data.drop(['Unnamed: 0'],axis=1)
```

```python
array=df.values
array
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
stscaler = StandardScaler().fit(array)
scaled_crime_data = stscaler.transform(array) 
scaled_crime_data
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(scaled_crime_data)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
clust=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
clust.head()
```

```python
crime_data_df = pd.concat([crime_data,clust],axis=1) 
crime_data_df.head(25)
```

```python
#East West Airlines DBSCAN Clustering
```

```python
EW_Airlines = pd.read_csv("EastWestAirlines.csv")
EW_Airlines.head()
```

```python
EW_Airlines1=EW_Airlines.iloc[:,1:]
```

```python
array=EW_Airlines1.values
array
```

```python
dbscan_a = DBSCAN(eps=1, min_samples=5)
dbscan_a.fit(scaled_Airlines)
```

```python
#Noisy samples are given the label -1.
dbscan_a.labels_
```

```python
clust_a
```

```python
Airlines_df = pd.concat([EW_Airlines,clust_a],axis=1) 
Airlines_df.head(25)
```
